```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
context_all_lbl_v7 = pd.read_csv("Data/version_7/context_all_lbl_v7.csv")
context_all_lbl_v7
```

```python
context_all_lbl_v7.urgency.value_counts()
```

```python
context_all_lbl_v7['urgency'] = np.where( (context_all_lbl_v7.urgency == 1) | (context_all_lbl_v7.urgency == 2), 0,
                                         np.where( (context_all_lbl_v7.urgency == 3), 1, 
                                                 np.where( (context_all_lbl_v7.urgency == 4) | (context_all_lbl_v7.urgency == 5), 2, np.NaN)) ).astype('int64')

context_all_lbl_v7.urgency.value_counts()
```

```python
context_all_lbl_v7.lang.value_counts()
```

```python
#select specific variables
data_df = context_all_lbl_v7[["mid","pid","sent_by","lang","urgency", "text", "system_context", "nurse_context"]]
data_df
```

```python
#split train, test, dev datasets
train_df, remain_df = train_test_split(data_df, random_state=42, train_size=0.7, stratify=data_df.urgency.values)
test_df, dev_df = train_test_split(remain_df, random_state=42, train_size=0.7, stratify=remain_df.urgency.values)
```

```python
train_df.shape, test_df.shape, dev_df.shape
```

```python
train_df.urgency.value_counts(), test_df.urgency.value_counts(), dev_df.urgency.value_counts()
```

```python
#generate datasets now
#system context
system_train_df_v7 = train_df[["mid","pid","sent_by","lang","urgency","system_context"]]
system_train_df_v7.rename(columns = {'system_context':'text', 'urgency':'label'}, inplace = True)
system_train_df_v7.to_csv("Experiment_44/input/system_train_df_v7.csv")
system_train_df_v7.to_csv("Experiment_45/input/system_train_df_v7.csv")

system_test_df_v7 = test_df[["mid","pid","sent_by","lang","urgency","system_context"]]
system_test_df_v7.rename(columns = {'system_context':'text', 'urgency':'label'}, inplace = True)
system_test_df_v7.to_csv("Experiment_44/input/system_test_df_v7.csv")
system_test_df_v7.to_csv("Experiment_45/input/system_test_df_v7.csv")

system_dev_df_v7 = dev_df[["mid","pid","sent_by","lang","urgency","system_context"]]
system_dev_df_v7.rename(columns = {'system_context':'text', 'urgency':'label'}, inplace = True)
system_dev_df_v7.to_csv("Experiment_44/input/system_dev_df_v7.csv")
system_dev_df_v7.to_csv("Experiment_45/input/system_dev_df_v7.csv")
```

```python
#nurse context
nurse_train_df_v7 = train_df[["mid","pid","sent_by","lang","urgency","nurse_context"]]
nurse_train_df_v7.rename(columns = {'nurse_context':'text', 'urgency':'label'}, inplace = True)
nurse_train_df_v7.to_csv("Experiment_46/input/nurse_train_df_v7.csv")
nurse_train_df_v7.to_csv("Experiment_47/input/nurse_train_df_v7.csv")

nurse_test_df_v7 = test_df[["mid","pid","sent_by","lang","urgency","nurse_context"]]
nurse_test_df_v7.rename(columns = {'nurse_context':'text', 'urgency':'label'}, inplace = True)
nurse_test_df_v7.to_csv("Experiment_46/input/nurse_test_df_v7.csv")
nurse_test_df_v7.to_csv("Experiment_47/input/nurse_test_df_v7.csv")

nurse_dev_df_v7 = dev_df[["mid","pid","sent_by","lang","urgency","nurse_context"]]
nurse_dev_df_v7.rename(columns = {'nurse_context':'text', 'urgency':'label'}, inplace = True)
nurse_dev_df_v7.to_csv("Experiment_46/input/nurse_dev_df_v7.csv")
nurse_dev_df_v7.to_csv("Experiment_47/input/nurse_dev_df_v7.csv")
```
